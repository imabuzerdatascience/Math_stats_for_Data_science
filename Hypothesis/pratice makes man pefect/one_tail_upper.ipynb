```python
import pandas as pd 
import numpy as np 
from scipy import stats
```

```python
population_mean = 50 
population_std = 10 
```

```python
one_tail_upper = pd.read_excel("z_test_one_tail_upper.xlsx" , sheet_name = "Data")
one_tail_upper.tail()
```

<h2> now we have to find sample mean and std </h2>

```python
sample_mean = one_tail_upper["Sample Data"].mean()
sample_std = one_tail_upper["Sample Data"].std()
sample_mean , sample_std
```

```python
std_error = population_mean/np.sqrt(one_tail_upper.shape[0])
std_error
```

```python
z_score = ( population_mean - sample_mean)/std_error
z_score
```

```python
Alpha = 0.05 
z_critical = stats.norm.ppf(1-Alpha)
z_critical 
```

```python
# compare two z_score and z_critical 
z_score > z_critical 
```

# Here z_score is less than z_critical so that why we reject the alternative hypothesis
