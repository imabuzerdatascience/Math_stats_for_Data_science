```python
import pandas as pd
import numpy as np 
from scipy import stats
```

```python
Population_Mean = 200
Population_Std_Dev = 20
```

```python
# import file 
two_tail_small = pd.read_excel("z_test_two_tail_small_sample.xlsx" , sheet_name = "Data" )
two_tail_small.tail()
```

```python
Alpha = 0.05 
```

```python
#  sample mean and std error 
sample_mean = two_tail_small["Sample Data"].mean()
std_error = Population_Mean / np.sqrt(two_tail_small.shape[0])

(sample_mean , std_error)
```

```python
# z_score 
z_score = (sample_mean - Population_Mean) / std_error
z_score
```

```python
z_critic = stats.norm.ppf(1-Alpha/2)
z_critic
```

```python
#  now compare 
z_score > z_score
```
