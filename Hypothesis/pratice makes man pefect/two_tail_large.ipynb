```python
import pandas as pd 
import numpy as np
from scipy import stats
```

```python
# popl mean 
population_mean = 500
#  std 
std = 60
```

```python
# import file 
two_tail_large = pd.read_excel("z_test_two_tail_large_sample.xlsx" , sheet_name = "Data" )
two_tail_large.tail()
```

```python
# sample mean 
sample_mean = two_tail_large["Sample Data"].mean()
std_error = population_mean / np.sqrt(two_tail_large.shape[0])

sample_mean , std_error
```

```python
z_score = (population_mean - sample_mean ) / std_error
z_score
```

```python
Alpha = 0.01 
```

```python
z_critic = stats.norm.ppf(1-Alpha/2)
z_critic
```

```python
#  now we compare the value 
z_score > z_critic
```

z_score is less than z_critic so we reject the alternative hypothesis
