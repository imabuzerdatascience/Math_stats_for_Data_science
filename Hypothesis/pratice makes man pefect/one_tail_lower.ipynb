```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#  population_mean 
population_mean = 100 
population_std = 15
```

```python
# load excel file 
one_tail_lower = pd.read_excel("z_test_one_tail_lower.xlsx" , sheet_name = "Data")
one_tail_lower.tail()
```

```python
# sample mean 
sample_mean = one_tail_lower["Sample Data"].mean()
sample_std_error = population_std / np.sqrt(one_tail_lower.shape[0])

sample_mean , sample_std_error
```

```python
z_score = ( population_mean - sample_mean) / sample_std_error
z_score
```

```python
Alpha = 0.05
```

```python
z_critic = stats.norm.ppf(1-Alpha)
z_critic
```

```python
# compare two z_score and z_critic 
z_score < z_critic 
```

```python
# z_score is less than z_critic so we reject the null hypothesis 
```
