```python
import pandas as pd 
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("house_price_increase.csv")
df
```

```python
df.shape
```

```python
df.price_increase_pct.describe()
```

```python
population_mean = 10
population_std = 4 
```

```python
sample_mean = df.price_increase_pct.mean()
sample_mean
```

```python
std_error = population_std / np.sqrt(df.shape[0])
std_error
```

```python
z_score = (sample_mean - population_mean)/std_error
z_score
```

```python
Alpha = 0.05
```

```python
z_critcal = stats.norm.ppf(1-Alpha)
z_critcal 
```

```python
z_score , z_critcal
```
