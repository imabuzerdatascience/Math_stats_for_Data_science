```python
import pandas as pd
import numpy as np
from scipy import stats 
```

```python
control_tst = pd.read_csv("control_group.csv")
control_tst.head()
```

```python
test_group = pd.read_csv("test_group.csv")
test_group.head()
```

```python
control_sample_mean = control_tst.recovery_time_hrs.mean()
control_std = control_tst.recovery_time_hrs.std()
control_shape = control_tst.shape[0] 
control_sample_mean , control_std , control_shape
```

```python
test_sample_mean = test_group.recovery_time_hrs.mean()
test_std = test_group.recovery_time_hrs.std()
test_shape = test_group.shape[0] 
test_sample_mean , test_std , test_shape
```

```python
A = control_std**2/control_shape
B = test_std**2/test_shape
z_score =(control_sample_mean - test_sample_mean)/np.sqrt(A+B) 
z_score
```

```python
Alpha = 0.05 
z_critical = stats.norm.ppf(1-Alpha)
z_critical 
```

```python
z_score > z_critical 
```

p_value 

```python
p_value = 1 - stats.norm.cdf(z_score)
p_value 
```

```python
p_value < Alpha 
```
