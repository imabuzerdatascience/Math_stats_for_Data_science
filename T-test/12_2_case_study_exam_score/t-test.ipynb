```python
import pandas as pd 
import numpy as np 
from scipy import stats
```

```python
df = pd.read_csv("exam_scores.csv")
df.head()
```

```python
population_mean = 72 
```

```python
sample_size = len(df)-1
```

```python
sample_mean = df.score.mean()
std = df.score.std(ddof=1)
std_error = std / np.sqrt(sample_size)
sample_mean  , std , std_error
```

```python
t_score = (sample_mean - population_mean ) / std_error
t_score
```

```python
Alpha = 0.05
```

```python
degree_of_freedom = sample_size
```

```python
t_critic = stats.t.ppf(1 - Alpha, degree_of_freedom)
t_critic
```

```python
# compare 
t_score > t_critic 
```
