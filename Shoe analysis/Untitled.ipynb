```python
# Convert 'date' to datetime format
def plot_qty():
    
   df['date'] = pd.to_datetime(df['date'])
   
   # Group data by date and brand, sum the sold quantities
   grouped = df.groupby(['date', 'brand'])['sold_qty'].sum().unstack()
   
   # Plot the line chart
   plt.figure(figsize=(12, 6))
   for brand in grouped.columns:
       plt.plot(grouped.index, grouped[brand], marker='o', label=brand)
   
   plt.title("Daily Shoe Sales by Brand")
   plt.xlabel("Date")
   plt.ylabel("Quantity Sold")
   plt.legend(title="Brand")
   plt.grid(True)
   plt.tight_layout()
   plt.xticks(rotation=45)
   plt.show()
```

```python
plot_qty()
```
