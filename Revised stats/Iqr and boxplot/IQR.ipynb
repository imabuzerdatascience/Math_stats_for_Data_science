```python
Q3 , Q1 = round(df.height.quantile([0.75 , 0.25 ]))
Q3 , Q1
```

```python
IQR = Q3 - Q1 
IQR
```

```python
Lower_bound = Q1 - 1.5 + IQR
Upper_bound = Q3 + 1.5 + IQR
Lower_bound , Upper_bound
```

### find outliers

```python
df[(df["height"] < Lower_bound ) | (df["height"] > Upper_bound )]
```

```python
val = round(df.height.median())
val
```

```python
df.height.replace( {1.2 : val , 40.2 : val } , inplace=True)
```
