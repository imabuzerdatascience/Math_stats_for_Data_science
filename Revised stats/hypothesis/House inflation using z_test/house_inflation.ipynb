```python
import pandas as pd 
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("house_price_increase.csv")
df.head()
```

```python
df.price_increase_pct.describe()
```

```python
population_mean = 10 
population_std = 4 
```

```python
sample_mean = df.price_increase_pct.mean()
std_error = population_std/np.sqrt(df.shape[0])
sample_mean , std_error 
```

```python
z_score = (sample_mean - population_mean ) / std_error 
z_score
```

```python
alpha = 0.05 
```

```python
z_critic = stats.norm.ppf(1-alpha/2).round(2)
z_critic 
```

```python
z_score , z_critic 
```

# p_value 

```python
p_value = 2*(1-stats.norm.cdf(abs(z_score)))
p_value , alpha
```
