```python
def plot_qty():
    plt.figure(figsize=(12 , 6))
    dates = df_adidas['date']
    plt.plot(dates , df_nike["sold_qty"] , marker="o" , color="blue" , label="Nike")
    plt.plot(dates, df_adidas['sold_qty'], marker='o', label='Adidas', color='red')
    plt.title("Daily Sales Qty for Nike and Adidas in September 2023")
    plt.xlabel("data")
    plt.ylabel("sold_qty")
    plt.legend()
    plt.xticks(rotation=45)
    plt.show()
plot_qty()
    
```
