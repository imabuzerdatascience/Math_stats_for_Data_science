```python
import pandas as pd 
from matplotlib import pyplot as pt 
import seaborn as sns
```

```python
df = pd.read_csv("heights_with_outliers.csv")
df.head(20)
```

```python
df.shape
```

```python
avg = df.height.mean()
std = df.height.std()
avg , std 
```

```python
sns.histplot(df.height , kde=True)
pt.show()
```

### now using 3std deviation to find that there is outliers r not 

```python
lower = round(avg - 3*std ,2 )
upper = round (avg + 3*std ,2 )
lower , upper 
```

```python
show_outliers = df[(df.height < lower ) | (df.height > upper) ]
show_outliers
```

```python
no_outliers = df[(df.height > lower ) & (df.height < upper) ]
no_outliers
```

```python
sns.histplot(no_outliers.height , kde= True)
```
