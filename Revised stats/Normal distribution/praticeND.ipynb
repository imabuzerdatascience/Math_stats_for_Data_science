```python
import pandas as pd 
from matplotlib import pyplot as pt 
import seaborn as sn 
```

```python
df_height = pd.read_excel("Height_Data_With_Outliers.xlsx")
df_height.tail()
```

```python
df_height.shape
```

```python
df_height.describe()
```

```python
mean = round(df_height.height.mean() ,2 )
std =  round(df_height.height.std() ,2 )
mean , std 
```

```python
sn.histplot(df_height.height , kde=True)
```

### finding data betn one std 

```python
left1 = mean - std 
right1 = mean + std 
left1 , right1
```

### Finfing data betn 2Std 

```python
left2 = mean - 2*std 
right2 = mean + 2*std 
left2 , right2
```

###        Find 3Std 

```python
left3 = mean - 3*std 
right3 = mean + 3*std 
left3 , right3
```

```python
find_outliers = df_height[(df_height.height < left3 ) | (df_height.height > right3 )]
find_outliers
```

```python
without_outliers_height = df_height[(df_height.height > left3 ) & (df_height.height < right3 )]
without_outliers_height
```

```python
sn.histplot( without_outliers_height.height , kde = True)
pt.show()
```
