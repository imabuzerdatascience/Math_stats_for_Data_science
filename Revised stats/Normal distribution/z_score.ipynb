```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv("heights_with_outliers.csv")
df.head(2)
```

```python
mean = round(df.height.mean() , 2 )
std = round (df.height.std() , 2 )
mean , std
```

```python
df["z_score"] = (df.height - mean )/std 
df.head()
```

```python
outliers_with_zscore = df[(df.z_score < -3 ) | (df.z_score > 3 )]
outliers_with_zscore
```

```python
without_outliers = df[(df.z_score > -3 ) & (df.z_score < 3 )]
without_outliers.head()
```

```python
sns.histplot(without_outliers.z_score)
```
