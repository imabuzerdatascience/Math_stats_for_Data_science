```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df_height = pd.read_csv("heights.csv")
df_height.tail()
```

```python
df_height.shape
```

```python
# find measure of dispersion to know that how the far the mean is from std 
mean = round(df_height.height.mean() , 2)
std = round(df_height.height.std() , 2 )
mean , std
```

```python
#plot histogram using sns 
sns.histplot(df_height , kde=True)
# show the mean in histogram 
plt.axvline(mean , color="black" , linestyle="dashed" , linewidth=2)

# one std 
plt.axvline(mean + std , color="r" , linestyle="dashed" , linewidth=2)
plt.axvline(mean - std , color="r" , linestyle="dashed" , linewidth=2)

# two std 
plt.axvline(mean + 2*std , color="g" , linestyle="dashed" , linewidth=2)
plt.axvline(mean - 2*std , color="g" , linestyle="dashed" , linewidth=2)

# Three std 
plt.axvline(mean + 3*std , color="b" , linestyle="dashed" , linewidth=2)
plt.axvline(mean - 3*std , color="b" , linestyle="dashed" , linewidth=2)


```

```python
left1 = mean - std
right1 = mean + std
left1 , right1
```

```python
left2 = mean - 2*std
right2 = mean + 2*std
left2 , right2
```

```python
left3 = mean - 3*std
right3 = mean + 3*std
left3 , right3
```

```python
# know that how much data is covered between left1 to right1 
df_height[(df_height.height > left1 ) & (df_height.height < right1 )].shape
```
