```python
import pandas as pd 
import matplotlib as pyplot
import seaborn as sns
```

```python
df = pd.read_csv("heights_with_outliers.csv")
df.head()
```

```python
sns.histplot(df)
```

```python
mean = round(df.height.mean() , 2)
std = round(df.height.std() , 2 )
mean , std
```

```python
# find lower and upper using 3 std 
lower = mean - 3*std
upper = mean + 3*std
lower , upper
```

```python
# now detect outliers 
df[(df.height < lower ) | (df.height > upper)]
```

```python
df_with_no_outliers = df[(df.height > lower ) & (df.height < upper)]
df_with_no_outliers
```

```python
sns.histplot(df_with_no_outliers , kde=True)
```
